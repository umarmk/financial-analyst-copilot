```python
import sys
import os

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
from src.metrics.core import (
    load_customer_month_mrr,
    load_revenue_events,
    get_mrr_by_month,
    get_mrr_components_by_month,
    get_net_new_mrr,
    get_revenue_churn_rate,
    get_active_customers,
)

month_mrr_df = load_customer_month_mrr()
events_df = load_revenue_events()
```

```python
mrr_by_month = get_mrr_by_month(month_mrr_df)
components = get_mrr_components_by_month(events_df)
net_new = get_net_new_mrr(components)
churn_rate = get_revenue_churn_rate(components, mrr_by_month)
active = get_active_customers(month_mrr_df)

mrr_by_month.head(), components.head(), net_new.head(), churn_rate.head(), active.head()
```

```python
# check how well Net New MRR matches actual MRR changes

merged = mrr_by_month.merge(net_new, on="month", how="left")
merged["mrr_change"] = merged["mrr_total"].diff()

merged[["month", "mrr_total", "mrr_change", "net_new_mrr"]].head(15)
```
